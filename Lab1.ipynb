Question 1 et 2 :

Remarque : on peut spécifier les probabilités d'apparition de 0 et 1 avec np.random.choice. D'autre part, plus
la taille des données est grande plus l'entropie tend vers 1

```python
#lab 1
#Question 1 et 2
import numpy as np
import math
import matplotlib.pyplot as plt
size=30 # à régler
tableau=np.random.choice(2, size, p=[0.5, 0.5]) # p contient les probabilités d'apparition de 0 et de 1 à spécifier
print("les données : ",tableau)
counts = np.unique(tableau, return_counts=True)
nbofones=counts[1][1]
nbofzeros=counts[1][0]
probabilityOfones=nbofones/size
probabilityOfzeros=nbofzeros/size
entropy=-probabilityOfones*math.log(probabilityOfones,2)-probabilityOfzeros*math.log(probabilityOfzeros,2)
print("l'entropie est : ",entropy) #if we choose a long size entropy will be more close to 1
```

### Question 5
Dans cette question, nous allons travailler sur l'image donnée sur moodle.
D'abord on va représenter l'image sous forme de matrice binaire (1 si pixel noir et 0 sinon)
Ensuite on va faire du comptage du nombre de 0 et de 1 et du nombre de 0 (resp 1) sachant qu'on a eu 1 (resp 0).
On obtient alors les 4 probabilités permettant d'afficher la matrice de transition.
Enfin, nous allons nous servir de la question 4 pour calculer les entropies avec et sans mémoire et les comparer

```python
import numpy as np
from PIL import Image
from scipy import *
image = Image.open('scan.png')
pix = np.array(image)
print("Voici la matrice qui représente l'image  de facon binaire (1 si noir et 0 sinon)")
print(pix)
counts = np.unique(pix, return_counts=True)
nbofones=counts[1][1]
nbofzeros=counts[1][0]
print("Nombre de zero  :",nbofzeros)
print("Nombre de un    :",nbofones)
nbZeroSachantZero=0
nbUnSachantUn=0
nbUnSachantZero=0
nbZeroSachantUn=0
for i in range(0,pix.shape[0]):
    for j in range(0,pix.shape[1]):    
        if(pix[i][j-1]==0 and pix[i][j]==0):
            nbZeroSachantZero+=1
        elif(pix[i][j-1]==1 and pix[i][j]==1):
            nbUnSachantUn+=1
        elif(pix[i][j-1]==0 and pix[i][j]==1):
            nbUnSachantZero+=1
        else:
            nbZeroSachantUn+=1

print("\nLe Nombre de zero sachant que on a eu zero avant est : ",nbZeroSachantZero)
print("Le Nombre de zero sachant que on a eu un avant est : ",nbZeroSachantUn)
print("Le Nombre de un sachant que on a eu zero avant est : ",nbUnSachantZero)
print("Le Nombre de un sachant que on a eu un avant est : ",nbUnSachantUn)
print("\nla proba d'avoir 0 sachant qu'on a eu 0 est : ",nbZeroSachantZero/nbofzeros)
print("la proba d'avoir 0 sachant qu'on a eu 1 est : ",nbZeroSachantUn/nbofones)
print("la proba d'avoir 1 sachant qu'on a eu 0 est : ",nbUnSachantZero/nbofzeros)
print("la proba d'avoir 1 sachant qu'on a eu 1 est : ",nbUnSachantUn/nbofones)
TransitionMatrix=zeros((2,2))
TransitionMatrix[0][0]=(nbZeroSachantZero)/nbofzeros
TransitionMatrix[0][1]=nbUnSachantZero/nbofzeros
TransitionMatrix[1][0]=nbZeroSachantUn/nbofones
TransitionMatrix[1][1]=nbUnSachantUn/nbofones
prob0=TransitionMatrix[1][0]/(1-TransitionMatrix[0][0]+TransitionMatrix[1][0])
prob1=1-prob0
print("\nla matrice de transition est donc : \n",TransitionMatrix)
print("\nRemarque : On voit que la proba d'avoir 1(resp 0) est tres élevée lorsqu'on a déja eu 1 (resp 0) ce qui est logique")
entropy=-(TransitionMatrix[0][0]*math.log(TransitionMatrix[0][0],2)+TransitionMatrix[0][1]*math.log(TransitionMatrix[0][1],2))*prob0-(TransitionMatrix[1][0]*math.log(TransitionMatrix[1][0],2)+TransitionMatrix[1][1]*math.log(TransitionMatrix[1][1],2))*prob1
print("l'entropie H(Sk) avec mémoire est : ",entropy)
print("\n\nP[x=0]= ",prob0)
print("P[x=1]= ",prob1)
print("l'entropie sans mémoire est donc : ",-prob0*math.log(prob0,2)-prob1*math.log(prob1,2))
print("\nRemarque : On voit que l'entropie sans mémoire est supérieure car on a moins d'informations ce qui est logique")
```

### Question 6
Dans cette question,j'ai d'abord fait une fonction d'encodage RLE et je l'ai testée sur une chaine de caractère
Ensuite j'ai appliqué cette fonction aux 220 premieres lignes de la matrice calculée précédemment ( le resultat sur toute l'image est trop long et illisible)

```python
#Question 6 encoder RLE
import re
from scipy import *
from PIL import Image
import numpy as np
image = Image.open('scan.png')
pix = np.array(image)
print("Voici la matrice qui représente l'image  de facon binaire (1 si noir et 0 sinon)")
print(pix)
ch=""
#RLE(run-length encoding) encoder
def encode(string):
    if string == '':
        return ''
    i = 0
    count = 0
    letter = string[i]
    rle = []
    while i <= len(string) - 1:        
        while string[i] == letter:
            i+= 1
            count +=1
            #catch the loop on last character so it doesn't got to top and access out of bounds
            if i > len(string) - 1:
                break
    
        rle.append('{1}{0}'.format(count, letter))
        if i > len(string) - 1: 
            break
        letter = string[i]
        count = 0
    #join list of strings together to create one string to return
    final = ''.join(rle)
    return final
print("\nTest de la fonction d'encodage sur la chaine: wwwwwwwwwwwbjwwwlpppzeeezzaaaaa\n",encode("wwwwwwwwwwwbjwwwlpppzeeezzaaaaa"))
print("\nApplication de RLE sur notre image de moodle (chaque ligne du tableau est transformée en chaine et encodée:\n")
print("Seul l'encodage des 220 premieres lignes est affiché")
for i in range(0,220):
    for j in range(0,pix.shape[1]):
        if pix[i][j]==0:
            ch+="0"
        else:
            ch+="1"
print(encode(ch))
```
